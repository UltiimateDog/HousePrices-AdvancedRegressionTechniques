#### Feature Agglomeration

```python
from sklearn.impute import SimpleImputer
from sklearn.cluster import FeatureAgglomeration

kept_features = ["SalePrice", "OverallQual", "Id"]
cluster = 8
DIR_DIM_REDUCTION = "DimensionReduction"

def agglomerate_housing_data(clusters, kept_features, id):
    housing_num = housing_numeric.drop(kept_features, axis=1)
    kept_feature = housing_numeric[kept_features].copy()
    
    imputer = SimpleImputer(strategy='mean')
    housing_num_imp = pd.DataFrame(imputer.fit_transform(housing_num), columns=housing_num.columns)
    
    agglo = FeatureAgglomeration(n_clusters=clusters)
    housing_num_agglo = agglo.fit_transform(housing_num_imp)
    
    housing_num_agglo_df = pd.DataFrame(housing_num_agglo)
    housing_num_agglo_df = pd.concat([housing_num_agglo_df, kept_feature], axis=1)
    corr_matrix = housing_num_agglo_df.corr()
    print(corr_matrix["SalePrice"].sort_values(ascending=False))
    housing_num_agglo_df.hist(bins=50, figsize=(20, 15))
    plt.savefig(os.path.join(DIR_DIM_REDUCTION, f"Agglo-{clusters}-{id}.png"))
    
    os.makedirs(DIR_DIM_REDUCTION, exist_ok=True)
    file_path = os.path.join(DIR_DIM_REDUCTION, f"Agglo-{clusters}-{id}.txt")
    with open(file_path, "w") as f:
        f.write(f"Feature Agglomeration\n")
        f.write(f"Clusters: {clusters}\n")
        f.write(f"Features kept: {kept_features}\n\n")
        f.write(str(corr_matrix["SalePrice"].sort_values(ascending=False)))
    
agglomerate_housing_data(cluster, kept_features, 3)
```

#### Principal Component Analysis

```python
from sklearn.decomposition import PCA

kept_features = ["SalePrice", "OverallQual", "Id"]
cluster = 8

def PCA_housing_data(clusters, kept_features, id):
    housing_num = housing_numeric.drop(kept_features, axis=1)
    kept_feature = housing_numeric[kept_features].copy()
    
    imputer = SimpleImputer(strategy='mean')
    housing_num_imp = pd.DataFrame(imputer.fit_transform(housing_num), columns=housing_num.columns)
    
    pca = PCA(n_components=clusters)
    housing_num_pca = pca.fit_transform(housing_num_imp)
    
    housing_num_pca_df = pd.DataFrame(housing_num_pca)
    housing_num_pca_df = pd.concat([housing_num_pca_df, kept_feature], axis=1)
    corr_matrix = housing_num_pca_df.corr()
    print(corr_matrix["SalePrice"].sort_values(ascending=False))
    housing_num_pca_df.hist(bins=50, figsize=(20, 15))
    plt.savefig(os.path.join(DIR_DIM_REDUCTION, f"PCA-{clusters}-{id}.png"))
    
    os.makedirs(DIR_DIM_REDUCTION, exist_ok=True)
    file_path = os.path.join(DIR_DIM_REDUCTION, f"PCA-{clusters}-{id}.txt")
    with open(file_path, "w") as f:
        f.write(f"PCA\n")
        f.write(f"Components: {clusters}\n")
        f.write(f"Features kept: {kept_features}\n\n")
        f.write(str(corr_matrix["SalePrice"].sort_values(ascending=False)))
        
PCA_housing_data(cluster, kept_features, 1)
```

#### Random Projections

```python
from sklearn.random_projection import SparseRandomProjection

kept_features = ["SalePrice", "OverallQual", "Id"]
cluster = 12


def SRP_housing_data(clusters, kept_features, id):
    housing_num = housing_numeric.drop(kept_features, axis=1)
    kept_feature = housing_numeric[kept_features].copy()

    imputer = SimpleImputer(strategy='mean')
    housing_num_imp = pd.DataFrame(imputer.fit_transform(housing_num), columns=housing_num.columns)

    srp = SparseRandomProjection(n_components=clusters)
    housing_num_srp = srp.fit_transform(housing_num_imp)

    housing_num_srp_df = pd.DataFrame(housing_num_srp)
    housing_num_srp_df = pd.concat([housing_num_srp_df, kept_feature], axis=1)
    corr_matrix = housing_num_srp_df.corr()
    print(corr_matrix["SalePrice"].sort_values(ascending=False))
    housing_num_srp_df.hist(bins=50, figsize=(20, 15))
    plt.savefig(os.path.join(DIR_DIM_REDUCTION, f"SRP-{clusters}-{id}.png"))

    os.makedirs(DIR_DIM_REDUCTION, exist_ok=True)
    file_path = os.path.join(DIR_DIM_REDUCTION, f"SRP-{clusters}-{id}.txt")
    with open(file_path, "w") as f:
        f.write(f"Sparse Random Projection\n")
        f.write(f"Components: {clusters}\n")
        f.write(f"Features kept: {kept_features}\n\n")
        f.write(str(corr_matrix["SalePrice"].sort_values(ascending=False)))

SRP_housing_data(cluster, kept_features, 2)
```

```python
def count_column_categories():
    count_array = []
    for column in range(len(housing_tr_cat_encoded[0])): ## iterates through columns
        temp_array = [0] * len(ordinal_encoder.categories_[column]) 
        for row in range(len(housing_tr_cat_encoded)): ## iterates row
            for category in range(len(ordinal_encoder.categories_[column])):
                if housing_tr_cat_encoded[row][column] == category:
                    temp_array[category] += 1
        count_array.append(temp_array)
    return count_array
```

```python
def print_categories_sum_and_save(filename="categories_sum.txt"):
    count_array = count_column_categories()
    with open(filename, "w") as file:
        for column in range(len(count_array)):
            file.write(f"\n{housing_tr_cat.columns.values[column]} ({sum(count_array[column])}):\n\n")
            for category in range(len(count_array[column])):
                file.write(f"\t{ordinal_encoder.categories_[column][category]}: {count_array[column][category]}\n")

# Call the function to save the output to "output.txt"
print_categories_sum_and_save()
```

```python
def filter_columns_by_non_null_count(df, threshold):
    """
    Filters the columns of a DataFrame based on the count of non-null values.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    threshold (int): The minimum number of non-null values required to retain a column.

    Returns:
    pd.DataFrame: A DataFrame with columns filtered by non-null count.
    """
    # Get the count of non-null values for each column
    non_null_counts = df.notnull().sum()

    # Filter the columns based on the threshold
    filtered_columns = non_null_counts[non_null_counts < threshold].index

    # Create a new DataFrame with the filtered columns
    filtered_df = df[filtered_columns]

    return filtered_df

housing_null_columns = filter_columns_by_non_null_count(housing_tr, len(housing_tr))
housing_null_columns.info()
```

```python
def print_null_percentage(df):
    """
    Prints the column names and the percentage of null values for each column in the DataFrame.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    """
    total_rows = len(df)
    null_percentages = df.isnull().sum() / total_rows * 100
    
    for column, null_percentage in null_percentages.items():
        print(f"{column}: {null_percentage:.2f}%")

print_null_percentage(housing_null_columns)
```

# Preparing data and transformation Pipelines

```python
housing_train = housing_tr.drop(["SalePrice","Id"], axis=1) ## id not necessary for training
housing_train_num = housing_train.select_dtypes(include='number')
housing_train_cat = housing_train.select_dtypes(exclude='number')
housing_labels = housing_tr["SalePrice"].copy()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class ExcludedAgglomeration(BaseEstimator, TransformerMixin):
    def __init__(self, features_to_keep_indices, n_clusters):
        self.features_to_keep_indices = features_to_keep_indices
        self.n_clusters = n_clusters
        self.agglomeration_ = None

    def fit(self, X, y=None):
        # Determine which features to agglomerate
        all_indices = np.arange(X.shape[1])
        features_to_agglomerate_indices = np.setdiff1d(all_indices, self.features_to_keep_indices)

        if len(features_to_agglomerate_indices) > 0:
            # Apply agglomeration to the features that are not kept
            self.agglomeration_ = FeatureAgglomeration(n_clusters=self.n_clusters)
            self.agglomeration_.fit(X[:, features_to_agglomerate_indices])
        return self

    def transform(self, X):
        # Check if agglomeration was applied
        if self.agglomeration_ is not None:
            all_indices = np.arange(X.shape[1])
            features_to_agglomerate_indices = np.setdiff1d(all_indices, self.features_to_keep_indices)

            if len(features_to_agglomerate_indices) > 0:
                # Apply agglomeration
                agglomerated_features = self.agglomeration_.transform(X[:, features_to_agglomerate_indices])
                
                # Combine kept features with agglomerated features
                return np.hstack([X[:, self.features_to_keep_indices], agglomerated_features])
            else:
                # If no features are to agglomerate, just return the kept features
                return X[:, self.features_to_keep_indices]
        else:
            # If no agglomeration was fit, just return the original data
            return X[:, self.features_to_keep_indices]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('agglomeration', ExcludedAgglomeration(features_to_keep_indices=[3] ,n_clusters=8)), ## 3 = OverallQual
    ('std_scaler', StandardScaler())
])

correlations = pd.concat([pd.DataFrame(num_pipeline.fit_transform(housing_train_num)), housing_labels], axis=1).corr()
correlations["SalePrice"].sort_values(ascending=False)
```

```python
class AllInOneCatTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, imputer_strategy, variance_threshold, n_components):
        self.imputer_strategy = imputer_strategy
        self.variance_threshold = variance_threshold
        self.n_components = n_components

    def fit(self, X, y=None):
        # Impute missing values
        self.imputer = SimpleImputer(strategy=self.imputer_strategy)
        X_imputed = self.imputer.fit_transform(X)
        
        # One-hot encode
        self.encoder = OneHotEncoder(handle_unknown='ignore')
        X_encoded = self.encoder.fit_transform(X_imputed)
        
        # Variance threshold
        self.var_thresh = VarianceThreshold(threshold=self.variance_threshold)
        X_var_thresh = self.var_thresh.fit_transform(X_encoded)
        
        # Truncated SVD
        self.svd = TruncatedSVD(n_components=self.n_components)
        self.svd.fit(X_var_thresh)
        
        return self

    def transform(self, X):
        # Impute missing values
        X_imputed = self.imputer.transform(X)
        
        # One-hot encode
        X_encoded = self.encoder.transform(X_imputed)
        
        # Variance threshold
        X_var_thresh = self.var_thresh.transform(X_encoded)
        
        # Truncated SVD
        X_reduced = self.svd.transform(X_var_thresh)
        
        return X_reduced
```

```python
cat_pipeline = Pipeline([
    ('all_in_one', AllInOneCatTransformer(imputer_strategy='most_frequent', variance_threshold= 0.16, n_components=25)),
])

test = pd.DataFrame(cat_pipeline.fit_transform(housing_train_cat))
test.describe()
```

#### Full pre process pipeline

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(housing_train_num)
cat_attribs = list(housing_train_cat)

preprocess_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', cat_pipeline, cat_attribs)
])

housing_prepared = preprocess_pipeline.fit_transform(housing_train)
housing_prepared
```

# Select and train Model

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(housing_prepared, housing_labels)
```

```python
some_data = housing_train.iloc[:5]
some_labels = housing_labels.iloc[:5]
some_data_prepared = preprocess_pipeline.transform(some_data)
print("Predictions: ", lin_reg.predict(some_data_prepared))
print("Labels: ", list(some_labels))
```

```python
from sklearn.metrics import mean_squared_error
housing_predictions = lin_reg.predict(housing_prepared)
lin_mse = mean_squared_error(housing_labels, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse ## -> $33,337 error 
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(housing_prepared, housing_labels)

housing_predictions = tree_reg.predict(housing_prepared)
tree_mse = mean_squared_error(housing_labels, housing_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse ## -> $153 error probably overfitting
```

#### Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, housing_prepared, housing_labels, 
                         scoring="neg_mean_squared_error",cv=10)
tree_rmse_scores = np.sqrt(-scores)

def display_scores(scores):
    print("Scores: ", scores)
    print("Mean: ", scores.mean())
    print("Standard deviation: ", scores.std())
    
display_scores(tree_rmse_scores)
"""
    Mean: 44,056
    Standard Deviation: 9,078
    Worse than linear regression
"""
```

## Random forest

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor(n_estimators=100, random_state=42)
forest_reg.fit(housing_prepared, housing_labels)

housing_predictions = forest_reg.predict(housing_prepared)
forest_mse = mean_squared_error(housing_labels, housing_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse ## -> $12,161
```

```python
scores = cross_val_score(forest_reg, housing_prepared, housing_labels,
                         scoring="neg_mean_squared_error",cv=10)
forest_rmse_scores = np.sqrt(-scores)
display_scores(forest_rmse_scores)
"""
    Mean: 31,614
    Standard Deviation: 6,876
"""
```

# Fine tuning model

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators': [80, 100, 120], 'max_features': [20, 25, 30, 34]},
]

forest_reg = RandomForestRegressor(n_estimators=100, random_state=42)

grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring="neg_mean_squared_error",
                           return_train_score=True)

grid_search.fit(housing_prepared, housing_labels)
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```
